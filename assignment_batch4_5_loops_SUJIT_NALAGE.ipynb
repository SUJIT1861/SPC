### 25. Given a dictionary, find the key with the highest value using a for loop

```python
fruits = {"Mango": 90, "Apple": 120, "Strawberry": 160}

max_key = None
max_value = 0
for key in fruits:
    if fruits[key] > max_value:
        max_value = fruits[key]
        max_key = key
print("Key with highest value:", max_key)
print("Highest value:", max_value)
```
